# 0. Introduction

This is the introductory notebook for the personal-co2-analysis project. This data analysis repository will be focused on the following.

## 0.1 The Issue

Climate change is rightfully presented as an urgent issue that demands our attention, however, it isn't clear what ordinary people can do to make a real difference and actually help the environment. Information on it is diffuse, and options that are marketed as environmentally friendly are often misleading.
In the face of all of this, what can we do? Changes should be demanded. But while those changes aren't here, what changes in a single person's lifestyle can make the most difference? And how much will it help? Is caring about this even worth it in the big picture? The project will focus on these questions regarding $CO_{2}$ emissions. While there are other greenhouse gases and metrics to consider, $CO_{2}$ is an important and often discussed one.

## 0.2 Project Scope
- **What will be covered in this project**: Calculating the carbon impact behaviors, actions and lifestyle changes of a single human being.
- **What will not be covered in this project**: Calculating greenhouse gases other than carbon dioxide, or other environmental impact metrics. Actions and practices made by government and corporations that a single person has little to no control over.
- **Time ranges**: To ensure results are up to date, we'll use the 2010-2023 range for all analyses unless otherwise specified.

## 0.4 The Notebooks
The analysis of the question will be structured like this in the notebooks.

- **0. Introduction**: Contextualization of the issue, explanation of the project and the datasets used.
- **1. Base Knowledge**: Before we get into the analysis, it's worth going over the topic of greenhouse gases and carbon dioxide. We will learn what $CO_{2}$ is and why we should pay attention to it.
- **2. $CO_{2}$ By Sector**: Here, we'll learn what sector exactly emits the most $CO_{2}$. We'll explore and datasets related to this and analyze trends to grasp the reality of carbon emissions.
- **3. Personal Actions**: Each personal behavior can be tied to a specific sector we've analyzed. We'll map those actions to their related sectors, and calculate, on average, how much that behavior emits $CO_{2}$. We'll also analyze a dataset related to per capita $CO_{2}$ emissions by sector in the world.
- **4. Lifestyle Changes**: After discovering and ranking the personal actions that emit the most $CO_{2}$, we'll go on to figure out which lifestyle changes actually affect the environment, $CO_{2}$-wise.
- **5. Conclusion**: All our findings will be presented succinctly here, with graphs to accompany it. We'll also deal with some questions that naturally will arise from this analysis. Does analysing how one person affects the environment even matter when most of the damage isn't done by the average person? We conclude the project here.
