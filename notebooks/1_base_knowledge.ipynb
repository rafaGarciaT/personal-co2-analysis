## 1.6 Division of Sectors

There are many ways to divide the global economy into sectors. We will be using the methodology used by the data source. The sectors are the following:

### 1.6.1 Electricity and Heat
Related to energy.

This sector includes emissions from:
- Electricity and heat production plants that burn fossil fuels. (Also includes $CH_{4}$ and $N_{2}O$ emissions.)
    - This includes main activity plants (energy/heat is produced for sale) and autoproducer plants (energy/heat is produced for own use).
- Other energy industries. (Also related to fossil fuels) (Also includes $CH_{4}$ and $N_{2}O$ emissions.)

### 1.6.2 Transportation
Related to energy.

This sector includes emissions from:
- Fossil fuel combustion for transportation. (Also includes $CH_{4}$ and $N_{2}O$ emissions.)

### 1.6.3 Manufacturing and Construction
Related to energy.

This sector includes emissions from:
- Fossil fuel combustion for manufacturing and construction. (Also includes $CH_{4}$ and $N_{2}O$ emissions.)

### 1.6.4 Building
Also related to energy.

This sector includes emissions from:
- Fossil fuel combustion for residential, commercial and public services. (Also includes $CH_{4}$ and $N_{2}O$ emissions.)

### 1.6.5 Other Fuel Combustions
Related to energy.

This sector includes emissions from:
- Fossil fuel combustion for agriculture, forestry and fishing, and other sectors. (Also includes $CH_{4}$ and $N_{2}O$ emissions.)

### 1.6.6 Fugitive Emissions
Related to energy.

This sector includes emissions from:
- Emissions that might occur during the extraction, processing and distribution of natural gas and oil systems. (Also includes $CH_{4}$ emissions.)

### 1.6.7 Industrial Processes
This sector includes emissions from:
- Cement production.

### 1.6.8 Bunker Fuels
This sector includes emissions from:
- Fuel combustions in aviation.
- Fuel combustions in shipping.

Here, only international aviation and shipping are included. Domestic aviation and shipping are included in the transportation sector.

### 1.6.9 Land-Use Change and Forestry
This sector includes emissions from:

- Land use total (forestland, net forest conversion, drained organic soils, and fires).
